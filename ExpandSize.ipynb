```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
import re
t = {}
lasts = -1
with open('result_sizes') as f:
    for l in f:
        m = re.search('size: (\d+)', l)
        if m != None:
            lasts = int(m.group(1))
        m = re.search('(\d+) ns/op', l)
        if m != None:
            t[lasts] = int(m.group(1))
            
t2 = {}
lasts = -1
with open('result_sizes2') as f:
    for l in f:
        m = re.search('size: (\d+)', l)
        if m != None:
            lasts = int(m.group(1))
        m = re.search('(\d+) ns/op', l)
        if m != None:
            t2[lasts] = int(m.group(1))
```

```python
df = pd.DataFrame(t, index=[0]).T
df2 = pd.DataFrame(t2, index=[0]).T
```
