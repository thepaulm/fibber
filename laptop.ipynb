```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
with open('results_0') as f:
    d0 = [l.split('\t')[2].split()[0] for l in f]
```

```python
with open('results_5') as f:
    d5 = [l.split('\t')[2].split()[0] for l in f]
```

```python
df0 = pd.DataFrame(d0)
```

```python
df5 = pd.DataFrame(d5)
```

```python
df0[0] = pd.to_numeric(df0[0])
df5[0] = pd.to_numeric(df5[0])
```

```python
df0 = df0.apply(lambda x: x / (1000 * 1000))
df5 = df5.apply(lambda x: x / (1000 * 1000))
```

```python
plt.style.use('ggplot')
```

```python
data0 = pd.DataFrame({"pad0 infinite": df0[0:6][0].values, "pad0 2": df0[6:12][0].values, "pad0 4": df0[12:18][0].values}, index=[1, 2, 4, 8, 24, 32])
data5 = pd.DataFrame({"pad5 infinite": df5[0:6][0].values, "pad5 2": df5[6:12][0].values, "pad5 4": df5[12:18][0].values}, index=[1, 2, 4, 8, 24, 32])
```

```python
data0.plot()
```

```python
data5.plot()
```

```python
pd.concat([data0, data5]).plot()
```
